```python
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.impute import SimpleImputer
#create a data frame from dataset file .csv
filepath=(r"C:\Users\bagew\Downloads\DataSci\dataset\DataPreprocessing.csv")
df=pd.read_csv(filepath)
print(f"dataframe=\n{df} \n type={type(df)}")

# function to find missing values in df isna()
print(f"NaN=\n{df.isna()}")

# SimpleImputer find replace missing values
imputer = SimpleImputer(strategy='mean')
#   dataframe[[column names]]
"""
Explanation of fit_transform
fit: This method calculates the necessary statistics (e.g., mean, standard deviation) from the training data. It learns the parameters needed for the transformation.
transform: This method uses the parameters calculated by the fit method to apply the transformation to the data.
"""
df[['Age','Income']] = imputer.fit_transform(df[['Age','Income']])
print(f"Output Df=\n{df} \n type=\n{type(df)}")
```
