```python
# -*- coding: utf-8 -*-
"""
Created on Sat Aug 24 12:37:49 2024

@author: bagew
"""

"""
1) Write a Python program to build SVM model to Cancer dataset. The
dataset is available in the scikit-learn library. Check the accuracy of model
with precision and recall.
"""
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report,accuracy_score

# load dataset
cancer_data=datasets.load_breast_cancer()
print("Cancer dataset=\n")
print(cancer_data)
print(f"Cancer data keys=\n{cancer_data.keys()}")
print(f"Cancer data Set Features names=\n{cancer_data.feature_names}")
print(f"Cancer data Set target names=\n{cancer_data.target_names}")

# train_test_split
x=cancer_data.data #features
y=cancer_data.target #target

x_train,x_test,y_train,y_test=train_test_split(x,y)
print(f"x_train shape={x_train.shape}")
print(f"x_test shape={x_test.shape}")
print(f"y_train shape={y_train.shape}")
print(f"y_test shape={y_test.shape}")

# standardize the data
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)

# Initialize and Train the SVM Model
svm_model=SVC(kernel='linear',random_state=42)
"""
SVC(kernel='linear', random_state=42) is a line of code in Python that sets up a Support Vector Machine (SVM) classifier to use for a machine learning task. Here's what each part does in simple terms:
SVC: This stands for "Support Vector Classifier." It's a tool that helps classify data into different categories.
kernel='linear': The kernel decides how the SVM will separate the data into different categories. 'linear' means it will try to separate the data with a straight line (or a flat plane in higher dimensions). This is good when the data is somewhat straightforward and can be separated easily.
random_state=42: This is like setting a specific "seed" for random numbers. It ensures that every time you run the code, you get the same results. The number 42 is just a commonly used number; you could use any number here.
Okay! Imagine you have a bunch of toys, some are cars, and some are blocks. You want to put all the cars on one side of a line and all the blocks on the other side.
When we use SVC(kernel='linear', random_state=42), we are telling the computer to find the best way to draw that line so that all the cars are on one side and all the blocks are on the other side.

SVC: This is like a smart helper that knows how to draw the line.
kernel='linear': This tells the helper to use a straight line.
random_state=42: This is like giving the helper the same starting instructions every time, so it always draws the line the same way.
So, we're using a helper that knows how to draw a straight line to separate our toys into two groups!
"""
svm_model.fit(x_train,y_train)

#prediction
y_pred=svm_model.predict(x_test)
print(f"Predicition=\n{y_pred}")

#evaluate model using classification Report 
print(f"Classification Report=\n{classification_report(y_test,y_pred)}")
#evaluate model using accuracy_score
accu=accuracy_score(y_test,y_pred)
print(f"accuracy of model is={accu*100}")
```
